```python
#converting to greyscale
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
Image.fromarray(gray)
```

```python
# applying binary threshold
ret,thresh = cv2.threshold(gray,102,255,cv2.THRESH_BINARY_INV) 
Image.fromarray(thresh)
```

```python
#dilating the  image
kernel = np.ones((15,20), np.uint8) 
img_dilation = cv2.dilate(thresh, kernel, iterations=1) 
Image.fromarray(img_dilation)
```

```python
#finding contours
ctrs, hier = cv2.findContours(img_dilation.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[1])
```

```python
ROI = []
for i, ctr in enumerate(sorted_ctrs): 
    # Get bounding box 
    x, y, w, h = cv2.boundingRect(ctr) 
    # Getting ROI 
    roi = img_cv[y:y+h, x:x+w] 
    # show ROI 
#     cv2.imshow('segment no:'+str(i),roi) 
    #cv2.waitKey(0) 
#     cv2.rectangle(img_cv,(x,y),( x + w, y + h ),(0,255,0),1) 
    display(Image.fromarray(roi))
    print("Width and Height of ROI are: {} and {}".format(w,h))
    if (w >= 30 and w < img_cv.shape[1]) and (h >= 30 and h < img_cv.shape[0]): 
#         print("Width and Height of ROI are: {} and {}".format(w,h))
#          cv2.imwrite('/home/rachit/Desktop/ocr/{}.png'.format(i), roi)
        ROI.append(roi)
```

```python
for roi in ROI: 
    display(Image.fromarray(roi))
```

```python
text = []
for roi in ROI:
#     roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    crop = Image.fromarray(roi)
    basewidth = 400
    wpercent = (basewidth/float(crop.size[0]))
    hsize = int((float(crop.size[1])*float(wpercent)))
    crop = crop.resize((basewidth,hsize), Image.ANTIALIAS)
    
    enhancer = ImageEnhance.Sharpness(crop)
    crop = enhancer.enhance(5.0)
#     crop = crop.convert('LA')
#     display(crop)
    text.append(pyTes.image_to_string(crop))
```

```python
import re 
expr1 = r"^[0-9]{2}[-|]{1}[0-9]{2}[-|]{1}[0-9]{4}$"
expr2 = r"^[0-9]{2}[-|]{1}[0-9]{2}[-|]{1}[0-9]{2}$"
pattern1 = re.compile(expr1)
pattern2 = re.compile(expr2)
count = 0
dates = []
for string in text:
    string = string.replace(' ', '')
    if pattern1.match(string) or pattern2.match(string):
        print(string)
        dates.append(string)
        count += 1
        if count == 3:
            break
```

```python
dates_json = {'Policy Date': dates[0], 'Period of Insurace - From': dates[2], 'Period of Insurance - To': dates[1]}
import json 
with open('dates.json', 'w') as fp:
    json.dump(dates_json, fp)
```
